```python
import pandas as pd
import numpy as np
import os
import matplotlib.pyplot as plt
```

```python
labelfile = "../dataset/phase_1_v3/train_labels.csv"

labels = pd.read_csv(labelfile)
labels
```

```python
def plot_histogram(data, column, xTitle):
    # Plot the histogram
    value_counts = data[column].value_counts()
    value_counts.plot(kind='bar')
    plt.xlabel(xTitle)
    plt.ylabel('Frequency')
    plt.title('Histogram of '+xTitle)
    plt.grid(False)
    plt.show()
```

```python
plot_histogram(labels, 'Direction', 'Direction')
plot_histogram(labels, 'Node', 'Nodes')
plot_histogram(labels, 'Type', 'Types')
```

```python
ew_labels = labels[labels['Direction']=='EW']

plot_histogram(ew_labels, 'Node', 'Nodes')
plot_histogram(ew_labels, 'Type', 'Types')
```

```python
ns_labels = labels[labels['Direction']=='NS']

plot_histogram(ns_labels, 'Node', 'Nodes')
plot_histogram(ns_labels, 'Type', 'Types')
```
